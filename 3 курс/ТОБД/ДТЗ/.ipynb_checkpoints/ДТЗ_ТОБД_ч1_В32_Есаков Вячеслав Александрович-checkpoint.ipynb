__Задание 1__. В файле `steam_games.json` находится информация об играх из магазина Steam. При помощи пакета `json` считайте файл. Выведите на экран количество игр жанра RPG, самым популярным языком которых является русский. "Язык является самым популярным" означает, что ранг этого языка в списке языков равен 0.

```python
with open('steam_games.json', 'r', encoding = 'utf-8') as read_file:
    steam_games = json.load(read_file)

lang_counter = 0


for cur_game in steam_games:
    lang = cur_game['languages']

    if 'RPG' in cur_game['genres']:
        for cur_lang in lang:
             if cur_lang['rank'] == 0 and 'Russian' in cur_lang['lang']:
                #print(cur_game['name'])
                lang_counter += 1
                break
        
print('Количество игр RPG, самым популярным языком которого является Русский:', lang_counter)
```

__Задание 4__. Преобразуйте столбец `original_price` в числовой. Удалите строки, в которых значение в данном столбце больше, чем величина 3 квартиля столбца. Проведите [min-max нормализацию](https://en.wikipedia.org/wiki/Feature_scaling#Rescaling_(min-max_normalization) столбцов `original_price` и `selled`. Постройте диаграмму рассеяния на основе двух этих столбцов. Сделайте цвет точек черным, размер точек - 5. Добавьте подписи осей (вертикальной и горизонтальной) и заголовок рисунка. Проведите прямую линию от точки (0, 1) до (1, 0). Сделайте эту линию красной и пунктирной.

```python
games_df['original_price'] = games_df['original_price'].str.replace('$', '', regex = False)
games_df['original_price'] = games_df['original_price'].str.replace(',', '.', regex = False)
games_df['original_price'] = games_df['original_price'].str.replace("''", '', regex = False)
```

```python
games_df['original_price'] = games_df['original_price'].astype(float)

quart_3 = games_df['original_price'].quantile(0.75)
rows = np.where(games_df['original_price'] > quart_3)[0]

games_df = games_df.drop(rows)

games_df.head()
```

```python
def MinMax(ser):
    return (ser - ser.min()) / (ser.max() - ser.min())
    

games_df['original_price'] = MinMax(games_df['original_price'])
games_df['selled'] = MinMax(games_df['selled'])
games_df.head()
```

```python
fig, ax = plt.subplots(figsize = (12, 8))

ax.scatter(y = games_df['selled'], x = games_df['original_price'], s = 5, c = 'black')
ax.plot((0, 1), (1, 0), c = 'r', ls = '--')

ax.set_title('Correlation between amount of sold games and their price')

ax.set_xlabel('Original Price')
ax.set_ylabel('Selled')

plt.show()
```

__Задание 5__. Для каждого описания из столбца `game_description` таблицы `games_df` при помощи регулярных выражений выясните, содержатся ли в нем символы '#', '[', ']', '“' или '%'. Добавьте 2 столбца: `has_special_symbols` и `n_special_symbols`. В столбце  `has_special_symbols` укажите True, если хотя бы один из указанных символов содержится в описании и False в противном случае. В столбце `n_special_symbols` укажите количество таких символов в строке.

```python
patt = re.compile('(\#|\[|\]|\"|\%)')

games_df['has_special_symbols'] = games_df.apply(lambda x: len(patt.findall(str(x['game_description']))) > 0, axis = 1)
games_df['n_special_symbols'] = games_df.apply(lambda x: len(patt.findall(str(x['game_description']))), axis = 1)

games_df.head()
```

__Задание 6__. Пересортируйте столбцы таблицы `games_df` по следующему правилу: сначала идут нечисловые столбцы, за ними идут числовые столбцы. Добавьте перед первым числовым столбцом пустой безымянный столбец. Сохраните текущее состояние `games_df` в рабочую книгу "games.xlsx". При помощи механизма группировки столбцов, который предоставляет Excel, создайте 2 группы столбцов: группу нечисловых столбцов и группу числовых столбцов.
Вставьте в ноутбук скриншоты листов Excel, подтверждающие корректность выполнения задания.

* Примечание 1: в интерфейсе Excel механизм группировки находится в разделе "Структура" на вкладке "Данные". Пустой столбец нужен для того, чтобы группы столбцов не сливались в одну.
* Примечание 2: вам может потребоваться преобразовать номер столбца в букву. Для этих целей можно воспользоваться функцией `xlsxwriter.utility.xl_col_to_name`.

```python
only_numeric = games_df.select_dtypes(include = [int, float, bool])
object_col = games_df.select_dtypes(include = [object])


total = object_col
total["Empty"] = ''
total = pd.concat([total, only_numeric], axis = 1)

total.head()
```

```python
first_cell = xlsxwriter.utility.xl_rowcol_to_cell(col = 0, row = 0)[0]
cell_before_empty = xlsxwriter.utility.xl_rowcol_to_cell(col = total.columns.get_loc('Empty'), row = 0)[0]
cell_after_empty = xlsxwriter.utility.xl_rowcol_to_cell(col = total.columns.get_loc('Empty') + 2, row = 0)[0]
last_cell = xlsxwriter.utility.xl_rowcol_to_cell(col = len(total.columns), row = 0)[0]
```

```python
w_book = xlsxwriter.Workbook('games_df.xlsx')

writer = pd.ExcelWriter('games_df.xlsx', engine = 'xlsxwriter')

total.to_excel(writer, sheet_name = 'games')
    
writer.sheets['games'].set_column(f'{first_cell}:{cell_before_empty}', None, None, {'level' : 1})
writer.sheets['games'].set_column(f'{cell_after_empty}:{last_cell}', None, None, {'level' : 1})

writer.close()
```
