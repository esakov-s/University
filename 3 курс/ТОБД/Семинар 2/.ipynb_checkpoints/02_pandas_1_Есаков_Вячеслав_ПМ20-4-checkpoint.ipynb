1.7 Измените индекс в таблице `reviews`, пронумеровав строки, начиная с нуля.

```python
reviews.index = np.linspace(0, len(reviews) - 1, len(reviews), dtype = int)
reviews.head()
```

```python
#Альтернативная версия решения - меняем первый столбец и переносим его в индексы, 
#так как при первом варианте ничего не меняется, а по первому заданию понятно, что значения там уникальные
reviews_copy = reviews.copy()


reviews_copy.index = reviews['id'].values
reviews_copy = reviews_copy.drop('id', axis = 1)
reviews_copy.head()
```

### Работа со строковыми данными в `pandas`

3.1  Добавьте в таблицу `recipes` столбец `description_length`, в котором хранится длина описания рецепта из столбца `description`.

```python
recipes['description_length'] = recipes['description'].str.len()
recipes.head()
```

3.2 Измените название каждого рецепта в таблице `recipes` таким образом, чтобы каждое слово в названии начиналось с прописной буквы.

```python
recipes['name'] = recipes['name'].str.title()

#recipes['name'] = recipes['name'][0].str.upper

recipes.head()
```

3.3 Добавьте в таблицу `recipes` столбец `name_word_count`, в котором хранится количество слов из названии рецепта (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

```python
recipes['name_word_count'] = recipes['name'].str.split().str.len()
```
