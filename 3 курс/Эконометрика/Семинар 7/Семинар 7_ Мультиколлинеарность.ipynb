##Алгоритм Фарара-Глоубера

Включает в себя 3 теста на мультикллинеарность, проверим по очереди каждый из них

####Хи квадрат статистика

```python
corr_matr = df.drop('Y', axis = 1).corr()
det = np.linalg.det(corr_matr.values)
det
```

```python
FG = -(17 - 11/6)*np.log(det)
FG
```

```python
hi_2 = sc.chi2.isf(0.05, 1/2 * 3 * 2)
hi_2
```

hi_2 (табл) > FG, следовательно мультиколлинеарность по данному тесту отсутствует

####F-тест

```python
obr = np.linalg.inv(corr_matr).diagonal()
obr
```

```python
F_stat = (obr - 1) * (14 / 3)
F_stat
```

```python
F_tabl = sc.f.ppf(0.95, 3, 14)
F_tabl
```

F_tabl > F_факт, в рамках F-теста у нас нет связи с другими регрессорами

##t-test

```python
c = np.linalg.inv(corr_matr)
r = np.copy(c)
for i in range(3):
  for j in range(3):
    r[i, j] = -c[i, j] / np.sqrt(c[i, i] * c[j, j])
r
```

```python
c = np.linalg.inv(corr_matr)
t = np.copy(r)
for i in range(3):
  for j in range(3):
    t[i, j] = abs(r[i, j] * np.sqrt(14 / (1 - r[i, j] ** 2)))
t
```

```python
t_tabl = sc.t.ppf(0.975, 14)
t_tabl
```

```python
t_tabl > t
```

t-tabl>все попарные значения t-теста для параметров(главная диагональ не рассматривается, следовательно по t-тесту мультиколлинеарность отсутствует
