# Создание модели данных и проверка гипотезы на них

```python
#Функция проверки критерия Колмогорова
def kolm_krit(n):
    data = np.random.normal(loc = 0, scale = 1, size = n) 
    result = stats.kstest(data,'norm')
    return data, round(result[1],3)
n = 252
```

```python
#построение гистограммы Р-значений, вычисленных с помощью критерия Колмогорова. (1 выборка)
from scipy.stats import kstest
P_values_1 = []
data_1 = kolm_krit(n)[0]
print(stats.kstest(data_1, 'norm'))

for i in range(1000):
    P_values_1.append(kolm_krit(n)[1])
    
plt.hist(P_values_1, rwidth=0.9)
plt.title('Гистограмма Р-значений 1 образца модельных данных, вычисленных на основе критерия Колмогорова')
plt.show()
```

```python
#построение гистограммы Р-значений, вычисленных с помощью критерия Колмогорова. (2 выборка)
from scipy.stats import kstest
P_values_2 = []
data_2 = kolm_krit(n)[0]
print(stats.kstest(data_2, 'norm'))

for i in range(1000):
    P_values_2.append(kolm_krit(n)[1])
    
plt.hist(P_values_2, rwidth=0.9)
plt.title('Гистограмма Р-значений 2 образца модельных данных, вычисленных на основе критерия Колмогорова')
plt.show()
```

```python
# Проверим, имеют ли сгенерированных 2 набора данных общий вид 
compar1 = stats.ks_2samp(data_1, data_2) 
print(compar1[1])
```

```python
#Функция проверки критерия Фишера
def fisher_crit(var1, var2):
    return ((max(var1, var2))**2/(min(var1, var2))**2)
```

```python
#Вычислим дисперсию 2 выборок данных
var_data_1 = np.var(data_1)
var_data_2 = np.var(data_2)
print("Дисперсия первой выборки = " + str(var_data_1), '\nДисперсия второй выборки = ' + str(var_data_2))

res_fisher = fisher_crit(var_data_1, var_data_2)
print("Критерий Фишера = ", res_fisher)
```

```python
#Проверка мощности критерия
m = 10000 #количество экспериментов
res1 = []
res2 = []
for i in range(m):
    data_11 = (stats.t.rvs(df = 3, loc = 0, scale = 1,size = 252))
    data_12 = (stats.t.rvs(df = 3, loc = 0, scale = 1,size = 252))
    data_21 = (stats.lognorm.rvs(s = 1/4 ,loc = 0, scale = 1,size = 252))
    data_22 = (stats.lognorm.rvs(s = 1/4 ,loc = 0, scale = 1,size = 252))
    var_data_11 = np.var(data_11)
    var_data_12 = np.var(data_12)
    var_data_21 = np.var(data_21)
    var_data_22 = np.var(data_22)
    res1.append(fisher_crit(var_data_11, var_data_12))
    res2.append(fisher_crit(var_data_21, var_data_22))

pow1 = 0
pow2 = 0
## Вычисление мощности критерия
for i in res1:
    if i < 0.05:
        pow1 += 1
print(pow1 / m)

for i in res2:
    if i < 0.05:
        pow2 += 1
print(pow2 / m)
```

## Проверка гипотезы на подготовленных реальных данных фондового рынка

```python
#импорт таблиц
import warnings
warnings.filterwarnings("ignore")

def tables(ticker):
    company = pd.read_csv(ticker+'.csv', sep=';') #чтение из файла
    company['Дата'] = company['Дата'].astype('object')
    for i in range(len(company['Дата'])):
        company['Дата'][i] = str(company['Дата'][i])
        if len(company['Дата'][i]) < 8:
            company['Дата'][i] = '0' + company['Дата'][i] 
    #print(company.info())
    company['Дата'] = pd.to_datetime(company['Дата'], format = '%d%m%Y') #преобразование даты
    #print(company)
    return company
# список логарифмов доходности по годам

def logYeld(ticker, num):
    LD = []
    ya = ['2016', '2017', '2018', '2019', '2020', '2021']
    company = tables(ticker)
    company['Логарифм доходности'] = np.log(company['Цена'])
    table = pd.read_csv('Общее количество торговых дней в году для выбранных компаний.csv', sep=';')
    table = table.drop(columns = table.iloc[:, range(1)]) 

    #print(table)
    for year in ya:
        
        
        n = (table[str(year)][num])
        #print(n)
        
        z = company['Логарифм доходности'][company['Дата'] <= str(year)+'-12-31'].size-(n)
        #print(z)
        LD.append((company['Логарифм доходности'][company['Дата'] <= str(year)+'-12-31'])[z:].kurtosis())
    
    print (LD)
    return LD


ticker_names = ['BSPB', 'CBOM', 'QIWIDR', 'SBER', 'SBER_p', 'SFIN', 'VTBR', 'MOEX']
PV = []
for cur_ticker in ticker_names:
    num = ticker_names.index(cur_ticker)
    log = logYeld(cur_ticker, num)
    
```

```python
def log_prof_max(file, year = 0): #cоздаю функцию для подсчета максимальной логарифмической доходности
    company = pd.read_csv(file, sep=';') #считываю файл
#обрезаю дату до четырех цифр года
    company['Дата'] = [int(str(company['Дата'][i])[4:]) for i in range(len(company['Дата']))]
#ищу логарифмическую доходность (т.е логарифм отношения цены закрытия текущей акции к предыдущей)
#округляю до двух знаков после запятой
    if year > 0:
        logg_doch_max = np.log(company['Цена'][company['Дата'] <= int(year)][company['Дата'] > int(year) - 1])
    else:
        logg_doch_max = np.log(company['Цена'])
    return [*stats.kstest(logg_doch_max, 'norm'), np.var(logg_doch_max)]
```

```python
#далее записываю в массив названия тикеров компаний для формирования столбца DataFrame и считывания файлов
all_suitable_tickers = ['BSPB', 'CBOM', 'QIWIDR', 'SBER', 'SBER_p', 'SFIN', 'VTBR', 'MOEX', 'MOEXFN']

#для формирования названий столбцов DataFrame создаю переменную, хранящую массив всех годов с 2016 по 2021
years = [i for i in range(2016,2022)]

#создаю DataFrame
table_log_max = pd.DataFrame() 

#записываю в столбец названия тикеров
table_log_max['Тикер'] = all_suitable_tickers
```

```python
#высчитываю максимальное отклонение цен по каждому году
all_val = []
for year in years:
    log_number_max = [] #создаю пустой массив для заполнения максимального отклонения по каждой акции
    for ticker in all_suitable_tickers:
#добавляю в массив значение макс.логарифмических доходностей
#по каждому году исследуемой акции из файла рассчитываю функции и добавляю в массив
        log_number_max.append(round(log_prof_max(ticker+'.csv', year)[1], 10))
        all_val.append(round(log_number_max[-1], 10))

#записываю в столбец по году значения массива log_number_max, который на каждом шаге цикла
#хранит максимальную логарифмическую доходность в определенный год по каждой акции
    table_log_max[str(year)] = log_number_max

#текст вывожу жирным
print('\033[1m' + 'Таблица. Р-значения по критерию Колмогорова')


table_log_max
```
