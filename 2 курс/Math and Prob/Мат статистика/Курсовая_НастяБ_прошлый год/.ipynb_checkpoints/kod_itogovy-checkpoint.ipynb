```python
#объем выборки
n=252 
#зададим функцию для вычисления статистики Шапиро-Уилка, для это используем функцию встроенную в Python
def Shapiro(n): 
    viborka_252 = np.random.normal(loc=0, scale=1, size=n) 
    znach_statistiki=stats.shapiro(viborka_252)[0]
    return znach_statistiki
```

```python
#вычисляем 1000 значений статистики Шапиро-Уилка
Shapir = [Shapiro(n) for m in range(1000)] 
```

```python
kvanlil_9 = [kv for kv in np.arange(0.1,1,0.1)]
znach_9=[]
for i in kvanlil_9:
    znach_9.append(np.quantile(Shapir,i))
df_9 = pd.DataFrame({'Квантиль':kvanlil_9, 'Значение':znach_9})
df_9.to_csv('Первые 9 квантилей статистики Шапиро-Уилка.csv',sep=';')
df_9
```

```python
kvanlil_999 = [kv for kv in np.arange(0.001,1,0.001)]
znach_999 = []
for i in kvanlil_999:
    znach_999.append(np.quantile(Shapir,i))
df_999 = pd.DataFrame({'Квантиль':kvanlil_999, 'Значение':znach_999})
df_999.to_csv('999 квантилей статистики Шапиро-Уилка.csv')
```

```python
#выполним проверку с помощью критерия Колмогорова
def p_kolm(n):
    viborka=np.random.normal(loc=0, scale=1, size=n) 
    p=stats.kstest(viborka,'norm')
    return round(p[1],3)
```

```python
#построение гистограммы Р-значений, вычисленных с помощью критерия Колмогорова
pvalues_kolm=[]
for i in range(1000):
    pvalues_kolm.append(p_kolm(n))
plt.hist(pvalues_kolm,rwidth=0.9)
plt.title('Гистограмма Р-значений модельных данных, вычисленных с помощью критерия Колмогорова')
plt.show()
```

```python
m = 10000 #количество экспериментов
shapir1=[]
shapir2=[]
for i in range(m):
    viborka1 = stats.t.rvs(df=1, loc = 0, scale = 1,size = 63)
    viborka2 = stats.lognorm.rvs(s=1 ,loc = 0, scale = 1,size = 63)
    shapir1.append(stats.shapiro(viborka1)[1])
    shapir2.append(stats.shapiro(viborka2)[1])
pow1 = 0
pow2 = 0
## Вычисление мощности критерия
for i in shapir1:
    if i<0.05:
        pow1+=1
print(pow1/m)
for i in shapir2:
    if i<0.05:
        pow2+=1
print(pow2/m)
```
