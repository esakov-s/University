```python
import scipy.stats as scs
from scipy import special, integrate
import numpy
import matplotlib.pyplot as plt
import math
import sympy as sym
```

Специальные дискретные случайные величины

```python
u1 = 1.9 ** 2
u2 = 1.8 ** 2
u3 = 1.5 ** 2


lambda_s = u1 + u2 + u3
p1 = scs.poisson.pmf(9, lambda_s)


print("Вероятность", p1)
print("Наиболее вероятное значение", lambda_s)
print("Матожидание", math.sqrt(lambda_s))
print("Ассиметрия", 1 / math.sqrt(lambda_s))
print("Эксцесс", 1 / lambda_s)
```

```python
n1 = 180
n2 = 70
p1 = 0.93
p2 = 0.92

    



e_s = n1 * p1 + n2 * p2
var_s = n1 * p1 * (1 - p1) + n2 * p2 * (1 - p2)
mu = n1 * p1 * (1 - p1) *(1 - 2*p1) + n2 * p2 * (1 - p2)*(1 - 2*p2)
print(math.sqrt(var_s))

print(mu/(var_s)**1.5)
```

Непрерывные случайные величины

```python
a = lambda x: (1 + 7 * x ** 0.5 + 8 * x ** 0.7 + 4 * x ** 0.9) ** 1.3
b = scipy.integrate.quad(a, 4, 9)
#print(b[0])
c = 1 / b[0]
print("Константа С:", c)
a = lambda x: c * (1 + 7 * x ** 0.5 + 8 * x ** 0.7 + 4 * x ** 0.9) ** 1.3

ax = lambda x: x * c * (1 + 7 * x ** 0.5 + 8 * x ** 0.7 + 4 * x ** 0.9) ** 1.3
e_s = scipy.integrate.quad(ax, 4, 9)
print("Математическое ожидание:", e_s[0])

ax2 = lambda x: (x - e_s[0]) ** 2 * c * (1 + 7 * x ** 0.5 + 8 * x ** 0.7 + 4 * x ** 0.9) ** 1.3
var_s = scipy.integrate.quad(ax2, 4, 9)
print("Дисперсия:", math.sqrt(var_s[0]))



#Попытка посчитать квантиль
#x = sym.Symbol('x')
#func = sym.integrate(c * (1 + 7 * x ** 0.5 + 8 * x ** 0.7 + 4 * x ** 0.9) ** 1.3, x)
#print(func)

#Квартиль формулами Насти
from scipy.stats import rv_continuous

class distr(rv_continuous):  
    def _pdf(self, x):
        return 0.0007986391189413239*(1+7.0*x**0.5+8.0*x**0.7+4.0*x**0.9)**1.3 if 4.0<=x<=9.0 else 0.0
    
X = distr()
print("Квантиль", X.ppf(0.9))
```

```python
#f = (1 + 6.0 * x**0.5 + 4.0 * x**0.7 + 5.0 * x**0.9)**1.3
def f(x):
    return (1 + 6.0 * x**0.5 + 4.0 * x**0.7 + 5.0 * x**0.9)**1.3

#print(f(4), f(8))


a = 4
b = 8
e_s = (a + b) / 2
print("Математическое ожидание:", f(e_s))





####
e_s2 = (a ** 2 + b ** 2 + a * b) / 3

var_s = math.sqrt((b - a) ** 2 / 12)


print(e_s, e_s2, var_s ** 2)
test = f(e_s2) - f(e_s ** 2)
print("Другой способ подсчёта", test)
print("Стандартное отклонение:", f(var_s))


ax2 = lambda x: (1 / (b - a)) * (1 + 6.0 * x**0.5 + 4.0 * x**0.7 + 5.0 * x**0.9)**1.3
var_s = scipy.integrate.quad(ax2, 4, 8)
print("Стандартное отклонение:", math.sqrt(var_s[0] - e_s ** 2))
```

```python
#Аналитическое решение
def f(x):
    return (1 + 6.0 * x**0.5 + 4.0 * x**0.7 + 5.0 * x**0.9)**1.3

x = []
y = []
i = 4.0000
while i <= 8.0000:
    x.append(i)
    y.append(f(i))
    i += 0.0001

plt.plot(x, y)
plt.show()

e_s = 0
for y in x:
    e_s += i * 1/(8 - 4)
print(e_s)
```
